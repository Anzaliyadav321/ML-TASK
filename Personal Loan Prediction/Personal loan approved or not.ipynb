```python
#convert categorical columns to numerical values
cat_cols = ['Gender', 'Home Ownership', 'Personal Loan' ]


for col in cat_cols:
     df[col] = pd.Categorical(df[col])

df = pd.get_dummies(df)   
```

```python
df.head()
```

```python
df.columns
```

```python
# dropping column
df.drop(['Personal Loan_ ','Gender_#', 'Gender_-'], axis=1, inplace=True)
df.head(5)
```

# people with 1 family member has taken more loan as compared with other 


```python
#plotting correlation heatmap
corr_mat = df.corr()

plt.figure(figsize=(20,12))
sns.heatmap(corr_mat, annot=True,cmap='coolwarm')
plt.title('correlation Heatmap')
plt.show()
```

```python
#combining Personal Loan_0 and Personal Loan_1 t0 Loan Approved 
df["Loan Approved"] = df["Personal Loan_0"] + df["Personal Loan_1"]
df.head()
```

```python
#dropping column
df.drop(['Personal Loan_0', 'Personal Loan_1'], axis=1, inplace=True)
```

```python
#seperating the data
X = df.drop(['Loan Approved'], axis=1)
Y = df['Loan Approved']
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Training and Evaluating the model

# Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, Y_train)
```

```python
Y_pred = random_forest.predict(X_test)
```

```python
# accuracy score on training data
from sklearn.metrics import accuracy_score
X_train_prediction = random_forest.predict(X_train)
training_data_accuray = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data : ', training_data_accuray)
```

```python
#accuracy on testing data
test_accuracy = accuracy_score(Y_pred, Y_test)
print('Accuracy on test data: ', test_accuracy)
```

# Support Vector Machine Model

```python
from sklearn.svm import SVC
support_vector_machine = SVC(kernel='linear')
support_vector_machine.fit(X_train, Y_train)
```

```python
Y_pred = support_vector_machine.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
#model saving
import joblib
joblib.dump(support_vector_machine, 'personal_loan_approval_model.pkl')
```

```python
#plotting confusion matrix
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(Y_test, Y_pred)
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(conf, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
